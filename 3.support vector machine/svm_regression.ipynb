```python
import pandas as pd
```

```python
dataset=pd.read_csv("RND.csv")
print(dataset)
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
print(dataset)
```

```python
dataset.columns
```

```python
independent=dataset[['R&D', 'Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_Newyork', 'State_newyork']]
print(independent)
```

```python
dependent=dataset[["Profit"]] 
print(dependent)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_train
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="rbf",C=100)
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_ #buying origin value
```

```python
regressor.n_support_  #how many model point get avaliable
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(x_test) 
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
print(r_score)
```
